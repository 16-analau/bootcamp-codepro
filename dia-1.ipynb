```python
#solicitar al usuario que ingrese el total de caramelos y niños
cantidad_de_caramelos = int(input('ingrese la cantidad de caramelos'))
cantidad_de_niños = int(input('ingrese la cantidad de niños'))
caramelos_recibidos = cantidad_de_caramelos / cantidad_de_niños
print(f'cada niño recibio {caramelos_recibidos} caramelos')

caramelos_sobrantes = cantidad_de_caramelos % caramelos_recibidos
print(f'sobran {caramelos_sobrantes} caramelos')
```
