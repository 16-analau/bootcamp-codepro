```python
#socitamos al usuario la edad para ingreso al asuncionico
edad_asuncionico = int(input('ingrese su edad para el acceso'))

#si tiene menos de 18

if edad_asuncionico <18:
    print('acceso denegado - solo zona kids')

elif edad_asuncionico <21:
    print('acceso permitido - no puede comprar bebidas alcoholicas')

else:
    print('acceso total - disfrute el festival')



```

```python
#cual es la temperatura del dia
temperatura_del_dia = int(input('ingrese la temperatura del dia'))


if temperatura_del_dia >=-10 and temperatura_del_dia <=13:
    print("ro'y koko")

elif temperatura_del_dia >=14 and temperatura_del_dia <=23:
    print("reresca'i")

elif temperatura_del_dia >=24 and temperatura_del_dia <=28:
    print('oikoite')

elif temperatura_del_dia >=29 and temperatura_del_dia <=45:
    print('hakuroooo')

else:
    print("ñamano mba'e")
```
